## URL Request

```python
import urllib.request
import time

class UrlFetcher(object):
    """ Implements the steps of fetching a URL.
        Main methods:
        fetch - Fetch the URL.
        get - Return the URLs data.
    """
 
    def __init__(self, url, timeout=30, retries=3, headers={}):
        """ Initializer.
        @params
         url - URL to fetch.
         timeout - Timeout per connection (seconds).
         entries - Max number of retries.
         headers - Optional request headers.
        """
    
        self.url = url
        self.timeout = timeout
        self.ntries = retries
        self.headers = headers
 
        # Enapsulated result object
        self.result = None

    def fetch(self):
        """ 
          Fetch the URL. Use configured timeouts, retries and
          header properties to customize fetch 
        """

        # This loop performs a network fetch of the URL, retrying
        # upto 'ntries' times in case of errors.

        count, result = 0, None
 
        while count < self.ntries:
            try:
                result = urllib.request.urlopen(self.url,
                    timeout=self.timeout,
                    headers = self.headers)
            except Exception as e:
                print('Caught error, trying again %s' % (e.args))
            # increment count
            count += 1
            time.sleep(1)
            if result != None:
                self.result = result
 
    def get(self):
        """ Return the data for the URL """
        if self.result != None:
            return self.result.content
```

```python
f = UrlFetcher('http://www.yahoo.com')
```

```python
f.fetch()
```

## Queue

```python
from queue import *
import threading
import time

def worker():
    while True:
        item = q.get()
        time.sleep(2)
        q.task_done()

num_worker_threads = 3
q = Queue()
for i in range(num_worker_threads):
     t = threading.Thread(target=worker)
     t.daemon = True
     t.start()

for item in range(20):
    q.put(item)

q.join()       # block until all tasks are done
```

## Map and Lambda

```python
from array import *

def rms(varray=[]):
    """ Root mean squared velocity. Returns
    square root of sum of squares of velocities """
    squares = map(lambda x: x*x, varray)
    return pow(sum(squares), 0.5)
```

## Graph Shortest Path

```python
# Shortest Path
# http://stackoverflow.com/questions/1323410/has-key-or-in

def find_path(graph, start, end, path=[]):
    path = path + [start]
    if start == end:
        return path
    if not start in graph:
        return None
    shortest = None
    for node in graph[start]:
        if node not in path:
            newpath = find_path(graph, node, end, path)
            if newpath:
                if not shortest or len(newpath) < len(shortest):
                    shortest = newpath
    return shortest
```

## Binary Tree

```python
class Node:
    def __init__(self, val):
        self.l_child = None
        self.r_child = None
        self.data = val

def binary_insert(root, node):
    if root is None:
        root = node
    else:
        if root.data > node.data:
            if root.l_child is None:
                root.l_child = node
            else:
                binary_insert(root.l_child, node)
        else:
            if root.r_child is None:
                root.r_child = node
            else:
                binary_insert(root.r_child, node)

def in_order_print(root):
    if not root:
        return
    in_order_print(root.l_child)
    print(root.data)
    in_order_print(root.r_child)

def pre_order_print(root):
    if not root:
        return        
    print(root.data)
    pre_order_print(root.l_child)
    pre_order_print(root.r_child)    
```

```python
r = Node(3)
binary_insert(r, Node(7))
binary_insert(r, Node(1))
binary_insert(r, Node(5))
```

```python
print("in order:")
in_order_print(r)

print("pre order")
pre_order_print(r)
```

## Timer

```python
def common_items(seq1, seq2):
    """ Find common items between sequences"""
    
    common = []
    for item in seq1:
        if item in seq2:
            common.append(item)
            
    return common
```

```python
from time import perf_counter as timer_func
from contextlib import contextmanager

@contextmanager
def timer():
    """Timing function to measure routines"""
    try:
        start = timer_func()
        yield
    except Exception as e:
        print(e)
        raise
    finally:
        end = timer_func()
        print('Time spent=>', 1000.0*(end-start),'ms.')
```

## Substring Sequence

```python
# brute force
def sub_string_brute(seq1, seq2):
    """ Substring brute force """
    
    subs = []
    for item in seq2:
        for parent in seq1:
            if item in parent:
                subs.append(item)
                
```

```python
def slices(s, n):
    return map(''.join, zip(*(s[l:] for i in range(n))))
```

```python
def sub_string(seq1, seq2):
    """ return substrings"""
    
    min_l, max_l = min(map(len,seq2)), max(map(len, seq2))
    sequences = {}
    
    for i in range(min_l, max_l+1):
        for string in seq1:
        # Create sub sequences
            sequences.update({}.fromkeys(slices(string, i)))
            
    subs = {}
    for item in seq2:
        if item in sequences:
            subs.append(item)
            
    return subs
```
