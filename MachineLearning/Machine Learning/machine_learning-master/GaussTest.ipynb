```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def getGaussElement(mu=0,sigma=1): 
    
    return np.random.normal(mu,sigma,1)[0]
```

```python
N=100

t = np.empty(N)

for i in range(N):
    t[i] = getGaussElement()
```

```python
plt.plot(t)
```
