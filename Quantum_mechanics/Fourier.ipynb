# Fourier series and expansion

## Taylor series and periodic functions

+ While Taylor expansions are accurate locally, the difference between a periodic function and a given Taylor polynomial gets larger the further away we get from the point *c* about which teh Taylor series is calculated
+ For periodic functions, including those that we find in quatum mechanics we use Fourier expansion

## Periodic functions

+ A periodic function has the following property:
$$ f\left(x\right)=f\left(x+T\right),\quad T>0 $$

+ The sine and cosine functions come to mind
+ They are periodic with a period, T = 2*&pi;*, such that for this period we have:
$$ \sin{kx}, \quad \cos{kx}, \quad {k}\in\mathbb{Z} $$
+ So, for this function of sine we have:
$$ f\left(x\right)=\sin{kx} \\ f\left(x+2\pi\right)=\sin{\left[{k\left(x+2\pi\right)}\right]} $$
+ We have the same for the cosine function

## Fourier series

+ The Fourier series is defined as:
$$ \frac{{a}_{0}}{2}+\sum _{ k=1 }^{ \infty  }{ \left[ { a }_{ k }\cos { \left( kx \right)  } +{ b }_{ k }\sin { \left( kx \right)  }  \right]  }  $$
+ The coefficients *a*<sub>0</sub>, *a*<sub>1</sub>, *a*<sub>2</sub>, ... and *b*<sub>1</sub>, *b*<sub>2</sub>, *b*<sub>3</sub>, ... are called the *Fourier coefficients*
+ If the series converges, it has a periodicity of 2*&pi;*

+ Given a Fourier series that converges on an interval, let's say [-*&pi;*,*&pi;*], then it represents a function on that interval
$$ f\left( x \right) =\frac { { a }_{ 0 } }{ 2 } +\sum _{ k=1 }^{ \infty  }{ \left[ { a }_{ k }\cos { \left( kx \right)  } +{ b }_{ k }\sin { \left( kx \right)  }  \right]  }  $$
+ The function *f* must be periodic outside of the interval

+ We must try and compute some of the Fourier coefficients
+ We'll start by integrating both sides with respect to *x* on the interval[-*&pi;*,*&pi;*]:
$$ \int _{ -\pi  }^{ \pi  }{ f\left( x \right)  } dx=\int _{ -\pi  }^{ \pi  }{ \frac { { a }_{ 0 } }{ 2 }  } dx+\int _{ -\pi  }^{ \pi  }{ \sum _{ k=1 }^{ \infty  }{ \left[ { a }_{ k }\cos { \left( kx \right)  } +{ b }_{ k }\sin { \left( kx \right)  }  \right]  }  } dx\\ \int _{ -\pi  }^{ \pi  }{ f\left( x \right)  } dx=\int _{ -\pi  }^{ \pi  }{ \frac { { a }_{ 0 } }{ 2 }  } dx+\sum _{ k=1 }^{ \infty  }{ \left[ { a }_{ k }\int _{ -\pi  }^{ \pi  }{ \cos { \left( kx \right)  } dx } +{ b }_{ k }\int _{ -\pi  }^{ \pi  }{ \sin { \left( kx \right)  } dx }  \right]  }  $$
+ This is done only assuming that we can change the order of integration and summation, which is usually NOT the case!
+ Now, for every *k*, being part of the positive integers (natural numbers, &#8469;), we have:
$$ \int _{ -\pi  }^{ \pi  }{ \cos { \left( kx \right)  } dx } =\frac { 1 }{ k } \sin { \left( kx \right)  } \overset { \pi  }{ \underset { -\pi  }{ | }  } =\frac { 1 }{ k } \left[ \sin { \left( k\pi  \right) - } \sin { \left( -k\pi  \right)  }  \right] =0\\ \int _{ -\pi  }^{ \pi  }{ \sin { \left( kx \right)  } dx } =-\frac { 1 }{ k } \cos { \left( kx \right)  } \overset { \pi  }{ \underset { -\pi  }{ | }  } =-\frac { 1 }{ k } \left[ \cos { \left( k\pi  \right) - } \cos { \left( -k\pi  \right)  }  \right] =0 $$
+ Therefor we have:
$$ \int _{ -\pi  }^{ \pi  }{ f\left( x \right)  } dx=\int _{ -\pi  }^{ \pi  }{ \frac { { a }_{ 0 } }{ 2 }  } dx\\ \int _{ -\pi  }^{ \pi  }{ f\left( x \right)  } dx=\frac { { a }_{ 0 } }{ 2 } x\overset { \pi  }{ \underset { -\pi  }{ | }  } ={ a }_{ 0 }\pi  $$
+ Now, we have the value of *a*<sub>0</sub>:
$$ { a }_{ 0 }=\frac { 1 }{ \pi  } \int _{ -\pi  }^{ \pi  }{ f\left( x \right)  } dx $$

+ For successive values of *a*<sub>k</sub> and *b*<sub>k</sub>, we will multiply both sides of the equation, first with the cosine and then with the sine function of *nx*, where *n* &isin; &#8469; (the natural numbers) and again assuming that we can indeed change the order of integration and summation:
$$ \int _{ -\pi  }^{ \pi  }{ f\left( x \right) \cos { \left( nx \right)  }  } dx=\int _{ -\pi  }^{ \pi  }{ \frac { { a }_{ 0 } }{ 2 } \cos { \left( nx \right)  }  } dx+\sum _{ k=1 }^{ \infty  }{ \left[ { a }_{ k }\int _{ -\pi  }^{ \pi  }{ \cos { \left( kx \right) \cos { \left( nx \right)  }  } dx } +{ b }_{ k }\int _{ -\pi  }^{ \pi  }{ \sin { \left( kx \right) \cos { \left( nx \right)  }  } dx }  \right]  }   $$

+ Without going in to the detail we have the following three findings:
$$ \int _{ -\infty  }^{ \infty  }{ \frac { { a }_{ 0 } }{ 2 } \cos { \left( nx \right)  }  } dx=0,\quad n\in \left\{ 1,2,3,\dots  \right\} \\ \int _{ -\infty  }^{ \infty  }{ \sin { \left( kx \right)  } \cos { \left( nx \right)  }  } dx=0,\quad k\in \left\{ 1,2,3,\dots  \right\} ,\quad n\in \left\{ 1,2,3,\dots  \right\} \\ \int _{ -\infty  }^{ \infty  }{ \cos { \left( kx \right)  } \cos { \left( nx \right)  }  } dx=0,\quad n\neq k,\quad k\in \left\{ 1,2,3,\dots  \right\} ,\quad n\in \left\{ 1,2,3,\dots  \right\} \\ \int _{ -\infty  }^{ \infty  }{ \cos { \left( kx \right)  } \cos { \left( nx \right)  }  } dx=\pi,\quad n=k,\quad k\in \left\{ 1,2,3,\dots  \right\} ,\quad n\in \left\{ 1,2,3,\dots  \right\}    $$
+ This reduces to:
$$ \int _{ -\pi  }^{ \pi  }{ f\left( x \right) \cos { \left( nx \right)  }  } dx={ a }_{ n }\pi  $$
+ By multiplying out by the sine of *nx* we get (similar to above):
$$ \int _{ -\pi  }^{ \pi  }{ f\left( x \right) \sin { \left( nx \right)  }  } dx={ b }_{ n }\pi  $$

### The Euler-Fourier equations

+ We now have the three equations (called the *Euler-Fourier equations*) for the Fourier coefficients:
$$ { a }_{ 0 }=\frac { 1 }{ \pi  } \int _{ -\infty  }^{ \infty  }{ f\left( x \right)  } dx\\ { a }_{ k }=\frac { 1 }{ \pi  } \int _{ -\infty  }^{ \infty  }{ f\left( x \right) \cos { \left( kx \right)  }  } dx,\quad k\in \left\{ 1,2,3,\dots  \right\} \\ b_{ k }=\frac { 1 }{ \pi  } \int _{ -\infty  }^{ \infty  }{ f\left( x \right) \sin { \left( kx \right)  }  } dx,\quad k\in \left\{ 1,2,3,\dots  \right\}  $$

+ We can now say that when a Fourier series converges on some interval, then it converges to some function *f*, where *f* is periodic outside the interval and the Fourier coefficients satisfy the three Euler-Fourier equations

### Example problem

+ Calculate the Fourier series for the square wave function (assuming *f* is periodic outside of the interval):
$$ f\left(x\right)=0, \quad x\in\left(-\pi,0\right] \\ f\left(x\right)=1, \quad x\in\left(0,\pi\right] $$

#### Solution

+ Because we have different functions of *f* on the interval for our Euler-Fourier equations, we will have to do the integration in two parts, so as to cover the integral
    + For *a*<sub>0</sub>:
    $$ { a }_{ 0 }=\frac { 1 }{ \pi  } \int _{ -\pi  }^{ 0 }{ f\left( x \right)  } dx+\frac { 1 }{ \pi  } \int _{ 0 }^{ \pi  }{ f\left( x \right)  } dx\\ { a }_{ 0 }=\frac { 1 }{ \pi  } \left( \int _{ -\pi  }^{ 0 }{ 0 } dx+\int _{ 0 }^{ \pi  }{ 1 } dx \right) \\ { a }_{ 0 }=\frac { 1 }{ \pi  } \left( 0+\pi  \right) =\frac { 1 }{ \pi  } \pi =1 $$
    + For *a*<sub>k</sub>:
    $$ { a }_{ k }=\frac { 1 }{ \pi  } \int _{ -\pi  }^{ 0 }{ f\left( x \right) \cos { \left( kx \right)  }  } dx+\frac { 1 }{ \pi  } \int _{ 0 }^{ \pi  }{ f\left( x \right) \cos { \left( kx \right)  }  } dx\\ { a }_{ k }=\frac { 1 }{ \pi  } \int _{ -\pi  }^{ 0 }{ \left( 0 \right) \cos { \left( kx \right)  }  } dx+\frac { 1 }{ \pi  } \int _{ 0 }^{ \pi  }{ \cos { \left( kx \right)  }  } dx\\ { a }_{ k }=\frac { 1 }{ k\pi  } \sin { \left( kx \right)  } \overset { \pi  }{ \underset { 0 }{ | }  } =0 $$
    + For *b*<sub>k</sub>:
    $$ b_{ k }=\frac { 1 }{ \pi  } \left[ \int _{ -\pi  }^{ 0 }{ f\left( x \right) \sin { \left( kx \right)  }  } dx+\int _{ 0 }^{ \pi  }{ f\left( x \right) \sin { \left( kx \right)  }  } dx \right] \\ b_{ k }=\frac { 1 }{ \pi  } \left[ \int _{ -\pi  }^{ 0 }{ \left( 0 \right) \sin { \left( kx \right)  }  } dx+\int _{ 0 }^{ \pi  }{ \sin { \left( kx \right)  }  } dx \right] \\ b_{ k }=-\frac { 1 }{ k\pi  } \left[ \cos { \left( kx \right)  }  \right] \overset { \pi  }{ \underset { 0 }{ | }  } \\ b_{ k }=-\frac { 1 }{ k\pi  } \left[ \cos { \left( k\pi  \right) -\cos { \left( 0 \right)  }  }  \right] ,\quad k\in \left\{ 1,2,3,\dots  \right\} \\ b_{ k }=-\frac { 1 }{ k\pi  } \left[ \cos { \left( k\pi  \right) -1 }  \right] \\ { b }_{ k }=0,\quad k\in \left\{ 2,4,6,\dots  \right\} \\ { b }_{ k }=\frac { 2 }{ k\pi  } ,\quad k\in \left\{ 1,3,5,\dots  \right\}  $$
        + We can write this differently:
        $$ { b }_{ 2k }=0,\quad k\in \left\{ 1,2,3,\dots  \right\} \\ { b }_{ 2k-1 }=\frac { 2 }{ \left( 2k-1 \right) \pi  } ,\quad k\in \left\{ 1,2,3,\dots  \right\}  $$

+ This makes the Fourier series:
$$ { F }_{ n }\left( x \right) =\frac { { a }_{ 0 } }{ 2 } +\sum _{ k=1 }^{ \infty  }{ \left[ { a }_{ k }\cos { \left( kx \right) +{ b }_{ k } } \sin { \left( kx \right)  }  \right]  } \\ { F }_{ n }\left( x \right) =\frac { 1 }{ 2 } +\sum _{ k=1 }^{ \infty  }{ { b }_{ k }\sin { \left( kx \right)  }  } \\ { F }_{ n }\left( x \right) =\frac { 1 }{ 2 } +\sum _{ k=1 }^{ \infty  }{ { b }_{ 2k-1 }\sin { \left[ \left( 2k-1 \right) x \right]  }  } \\ { F }_{ n }\left( x \right) =\frac { 1 }{ 2 } +\sum _{ k=1 }^{ \infty  }{ \frac { 2 }{ \left( 2k-1 \right) \pi  } \sin { \left[ \left( 2k-1 \right) x \right]  }  } \\ { F }_{ n }\left( x \right) =\frac { 1 }{ 2 } +\frac { 2 }{ \pi  } \sin { x } +\frac { 2 }{ 3\pi  } \sin { 3x } +\frac { 2 }{ 5\pi  } \sin { 5x } +\dots  $$

+ Let's draw this for *n* = 7

```python
x = linspace(-2 * pi, 2 * pi, 1000)

plt.figure(figsize = (10, 8))
plt.plot(x, (1 / 2) + ((2 / pi) * sin(x)) + ((2 / 3 * pi) * sin(3 * x)) + ((2 / 5 * pi) * sin(5 * x)) + ((2 / 7 * pi) * sin(7 * x)), 'r-')
plt.show();
```

## The Fourier transform and inverse Fourier transform

+ The unitary, ordinary-frequency Fourier transform of a function *f* is defined as:
$$ F\left( k \right) =\int _{ -\infty  }^{ 
\infty  }{ f\left( x \right) { e }^{ -2\pi ixk } } dx $$

```python
from sympy import fourier_transform, exp
from sympy.abc import x, k
```

```python
f = exp(-x ** 2)
f
```

```python
fourier_transform(f, x, k)
```
